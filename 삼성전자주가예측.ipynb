```python
#네이버 금융 페이지에서 삼성전자 주식데이터 일별 시세를 크롤링한다.
#라이브러리를 임포트하고 html을 가져온다.
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

url = "https://finance.naver.com/item/sise_day.naver?code=005930&page=1"
url = requests.get(url)
print(url.text)
```

```python
#error가 떠서 크롤링을 방지하고 있음을 알 수 있다.
#따라서 요청시 헤더 정보를 user-agent로 변경하여 우회하여 가져온다.
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from tqdm import tqdm

# headers 정보 설정
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

total = []

for n in tqdm(range(1, 680)):
    
    # URL 설정
    url = "https://finance.naver.com/item/sise_day.naver?code=005930&page=1"

    # headers 정보와 함께 요청
    response = requests.get(url, headers=headers)

    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')
    soup = soup.find('table', class_='type2')
    table = pd.read_html(str(soup))[0]
    table = table.dropna() #NaN으로 표기된 결측치 제거

    total.append(table)
    time.sleep
```

```python
#KNN 선형회귀 사용
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
import numpy as np
from tensorflow import keras
```

```python
df = pd.read_excel('samsung.xlsx')
df
```

```python
#분석에 필요한 데이터 컬럼만 가져온다
data = df[['종가', '시가', '고가', '저가', '거래량']]
data
```

```python
#예측할 데이터는 종가이므로 종가 데이터를 target으로 지정
target = df['종가']
target
```

```python
#data는 마지막 날짜는 포함하지 않도록 하고, target은 첫번째 것은 빼고 가져온다
data = data[:-1]
target = target[1:]
data
target
```

```python
print(len(data), len(target))
```

```python
#분석을 위해 데이터를 행렬처리 해준다
data = data.to_numpy()
target = target.to_numpy()
data
```

```python
#데이터를 train 데이터와 test용 데이터로 분할
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
#분할한 train과 test 데이터에 대해 스케일링을 하고, knn을 모델을 학습시키고 성능 확인
mean = np.mean(x_train, axis=0)
std = np.mean(x_train, axis=0)
x_train_scaled = (x_train - mean) / std
x_test_scaled = (x_test - mean) / std

model = KNeighborsRegressor()
model.fit(x_train_scaled, y_train)
model.score(x_test_scaled, y_test)
```

```python
#linear regression으로 비교
model = LinearRegression()
model.fit(x_train_scaled, y_train)
model.score(x_test_scaled, y_test)
```

```python
dfx = df[['종가', '시가', '고가', '저가', '거래량']]
dfy = df['종가']
```

```python
x = dfx.values.tolist()
y = dfy.values.tolist()
x
```

```python
y
```

```python
#데이터 표준화 스케일링
data_x = []
data_y = []
window = 10

mean = np.mean(x, axis=0)
std = np.std(x, axis=0)

x = (x - mean) / std

#데이터를 10개 단위로 경향성을 줘서 rnn 학습을 시킨다
for i in range(len(y) - window):
    if y[i + window - 1] < y[i + window]:
        data_y.append(1)
    else:
        data_y.append(0)
    data_x.append(x[i:i+window])

data_x = np.array(data_x)
data_y = np.array(data_y)

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
model = keras.Sequential()

model.add(keras.layers.SimpleRNN(10, activation='relu', input_shape=(10,5)))
model.add(keras.layers.Dropout(0,1))
model.add(keras.layers.Dense(1, activation='sigmoid'))

stop = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs=1000, callbacks=[stop])
```

```python
model.evaluate(x_test, y_test)
```

```python
#simplernn 말고 lstm으로도 진행해본다.
#데이터 표준화 스케일링
data_x = []
data_y = []
window = 10

mean = np.mean(x, axis=0)
std = np.std(x, axis=0)

x = (x - mean) / std

#데이터를 10개 단위로 경향성을 줘서 lstm 학습을 시킨다
for i in range(len(y) - window):
    if y[i + window - 1] < y[i + window]:
        data_y.append(1)
    else:
        data_y.append(0)
    data_x.append(x[i:i+window])

data_x = np.array(data_x)
data_y = np.array(data_y)

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
model = keras.Sequential()

model.add(keras.layers.LSTM(10, activation='relu', input_shape=(10,5)))
model.add(keras.layers.Dropout(0,1))
model.add(keras.layers.Dense(1, activation='sigmoid'))

stop = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs=1000, callbacks=[stop])
```

```python
model.evaluate(x_test, y_test)
```
